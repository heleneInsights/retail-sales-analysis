## Data Cleaning

### Formatting `Date` column

To preserve the original dataset, a copy of the DataFrame is created. 
Unlike simple assignment, which creates a reference to the same object, the `.copy()` method creates an independent DataFrame in memory.

The `Date` column is converted to a datetime data type using pandas, enabling temporal operations such as filtering and sorting.

Invalid or non-parseable values are coerced into `NaT` to prevent errors and allow identification of problematic entries.

It is now possible to apply temporal filters based on the `Date` column.

No missing values were found in the dataset after validation using `.isna().sum()`.  

Therefore, no further action was required for handling missing data.

````python
df_clean = df.copy()
````

Check if the conversion ended successfully.
Invalid dates become `NaT` instead of crashing.

````python
df_clean["Date"] = pd.to_datetime(df_clean["Date"], errors="coerce")
df_clean["Date"].isna().sum()
````
