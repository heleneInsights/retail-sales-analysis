## Analysis

### Core Metrics

#### Total Sales per Product Category

Pandas was used for data selection and filtering, while NumPy was used to perform aggregation operations.

A short function was implemented to improve the readability of the results.

Sales are relatively evenly distributed across categories, with each contributing between 31% and 34% of total revenue. Electronics shows the highest share, while Beauty contributes the least, although the differences are small.

````python
categories = df_analysis["Product Category"].unique()
````

````python
sales_per_category =  {}

for category in categories:
    condition = df_analysis["Product Category"] == category
    total_sales = np.sum(df_analysis.loc[condition, "Total Amount"])
    sales_per_category[category] = total_sales
````

````python
def format_number(n):
    """
    Replace thousand separator by . and decimal separator by ,
    """
    return f"{n:,.2f}".replace(",", "X").replace(".", ",").replace("X", ".")
````

````python
def sorted_dict(messy_dict: dict):
    """
    Sort a dict from highest to lower value.
    """
    return sorted(
        messy_dict.items(),
        key=lambda x: x[1],
        reverse=True
        )
````

````python
sorted_sales_per_categories = sorted(
    sales_per_category.items(),
    key=lambda x: x[1],
    reverse=True
)
````

````python
max_len = max(len(cat) for cat in categories) + 4
sorted_sales_per_categories = sorted_dict(sales_per_category)

print("\nTotal Sales per Product Category\n")
print(f"{'Category':<{max_len}}{'Total Sales':>17}")
print("-" * (max_len + 17))

for category, total_sales in sorted_sales_per_categories:
    print(f"{category:<{max_len}}{format_number(total_sales):>17}")
````

````python
total_sales_all = sum(sales_per_category.values())

print("\nCategory Contribution (%)\n")
print(f"{'Category':<{max_len}}{'% Contribution':>17}")
print("-" * 32)

for category, total_sales in sorted_sales_per_categories:
    percentage = (total_sales / total_sales_all) * 100
    print(f"{category:<{max_len}}{format_number(percentage):>16}%")
````

#### Average Daily Sales per Product Category

Average daily sales per category follow the same pattern as total sales, with minimal variation between categories. This indicates that revenue is evenly distributed across categories, although this metric does not capture variability between individual days.

While average daily sales are stable across categories, transaction counts vary across days, ranging from 1 to 11 purchases.

````python
average_daily_sales = {}

num_days = df_analysis["Date"].nunique()

for category, total_sales in sales_per_category.items():
    avg = total_sales / num_days
    average_daily_sales[category] = avg
````

````python
sorted_averages = sorted_dict(average_daily_sales)

print("\nAverage Daily Sales per Product Category\n")
print(f"{'Category':<{max_len}}{'Avg Daily Sales':>25}")
print("-" * (max_len + 25))

for category, avg in sorted_averages:
    print(f"{category:<{max_len}}{format_number(avg):>25}")
````

````python
daily_transactions = df_analysis["Date"].value_counts()

print("\nTop 5 Days by Transaction Count\n")
print(daily_transactions.head())
````

````python
min_tx = daily_transactions.min()
max_tx = daily_transactions.max()

print(f"\nTransaction counts range from {min_tx} to {max_tx} purchases.")
````

### Addional Metrics

#### Average Order Value per Category

The Average Order Value (AOV) was analyzed across product categories and further segmented by gender and age group to better understand customer behavior.

Initial results showed relatively small differences in AOV across categories, indicating that product category alone is not a strong differentiator of customer spending.

When incorporating demographic segmentation, the Electronics · Male · 51–65 segment emerged as the most valuable, combining high average order value with strong total revenue contribution.

Further analysis identified underperforming segments with high potential. In particular, the Clothing · Female · 18–25 segment stands out: despite generating relatively low revenue and order volume, it exhibits one of the highest AOVs.

This suggests that customers in this segment are willing to spend more per purchase but are not purchasing frequently, representing an opportunity to increase revenue by improving engagement and conversion.

A similar pattern is observed in the Electronics · Female · 18–25 segment, though with slightly higher activity and therefore lower relative upside.

````python
df_analysis['Age Group'] = pd.cut(
    df_analysis['Age'],
    bins=[18, 25, 35, 50, 65],
    labels=['18-25', '26-35', '36-50', '51-65']
)
````

````python
segment_metrics = (
    df_analysis.groupby(['Product Category', 'Gender', 'Age Group'])
        .agg(
            total_revenue=('Total Amount', 'sum'),
            avg_order_value=('Total Amount', 'mean'),
            orders=('Total Amount', 'count')
        )
    .reset_index()
)
````

````python
segment_metrics['value_score'] = (
    (segment_metrics['total_revenue'] / segment_metrics['total_revenue'].max()) * 0.6 +
    (segment_metrics['avg_order_value'] / segment_metrics['avg_order_value'].max()) * 0.3 +
    (segment_metrics['orders'] / segment_metrics['orders'].max()) * 0.1
)
````

````python
low_revenue_threshold = segment_metrics['total_revenue'].quantile(0.30)
low_orders_threshold = segment_metrics['orders'].quantile(0.30)
high_aov_threshold = segment_metrics['avg_order_value'].quantile(0.70)

underperforming_high_potential = segment_metrics[
    (segment_metrics['total_revenue'] < low_revenue_threshold) &
    (segment_metrics['orders'] < low_orders_threshold) &
    (segment_metrics['avg_order_value'] > high_aov_threshold)
]

# Minimum orders
underperforming_high_potential = underperforming_high_potential[
    underperforming_high_potential['orders'] >= 5
]
````

````python
# Top segments (business value)
top_segments = segment_metrics.sort_values(by='value_score', ascending=False).round(2)

top_segments
````

**Recommendation:**

> Focus on retaining and maximizing these high-value segments through loyalty programs, personalized offers, and premium product positioning, as they are the primary drivers of revenue.

````python
# Opportunity segments
opportunity_segments = underperforming_high_potential.sort_values(
    by='avg_order_value', ascending=False
).round(2)

opportunity_segments
````

**Recommendation:**

> Increase purchase frequency (used here as a proxy for customer engagement) in these segments through targeted marketing campaigns and tailored promotions, as customers already demonstrate strong spending potential.

#### Total Quantity Sold per Category

Clothing shows the highest sales volume, with Electronics close behind, indicating relatively consistent demand across these categories. Beauty has the lowest quantity sold, suggesting comparatively lower demand or a more niche positioning.

Despite these differences in volume, revenue is evenly distributed across all product categories, with each contributing roughly one-third of total sales. This indicates a balanced business without over-reliance on a single category.

````python
df_analysis.groupby('Product Category')['Quantity'].sum()
````

````python
df_analysis.groupby('Product Category')['Total Amount'].sum()
````

To better compare performance across categories, quantity, revenue, and average order value were combined into a single summary table:

````python
category_summary = (
    df_analysis.groupby('Product Category')
        .agg(
            total_quantity=('Quantity', 'sum'),
            total_revenue=('Total Amount', 'sum'),
            avg_order_value=('Total Amount', 'mean')
        )
        .reset_index()
)

# Add revenue contribution
category_summary['revenue_share_%'] = (
    category_summary['total_revenue'] / category_summary['total_revenue'].sum()
) * 100

category_summary = category_summary.round(2)
````

````python
category_summary.sort_values(by='revenue_share_%', ascending=False)
````

**Insights:**

> All categories contribute similarly to total revenue, with no clear dominant category. Clothing leads in volume, while Beauty shows slightly higher average order value, indicating different performance drivers across categories.

**Recommendation:**

> Maintain a balanced strategy across categories while optimizing their individual strengths: support volume in Clothing, improve demand generation in Beauty, and sustain the balanced performance of Electronics.
