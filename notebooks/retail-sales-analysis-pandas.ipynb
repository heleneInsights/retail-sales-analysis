We also checked if the word "Manzana" appears in any text column.

```python
# Check if "Manzana" appears in any string column
str_cols = df.select_dtypes(include="str")
```

```python
for col in str_cols.columns:
    if str_cols[col].str.contains("Manzana", case=False, na=False).any():
        print(f"'Manzana' found in column: {col}")
```

## Data Filtering

We applied different conditions to filter the dataset and explore specific cases.

Since `Ventas` and `Precio` were not available, `Total Amount` and `Price per Unit` were used as their equivalents for filtering.

The dataset does not contain a product named `Manzana`, so this condition could not be applied. Instead, a similar filter was tested using an available category (`Electronics`) with a higher sales condition.

```python
condition = df_clean['Total Amount'] > 50

df_clean[condition]
```

```python
condition = df_clean['Price per Unit'] < 0.5

df_clean[condition]
```

```python
condition = "(`Product Category` == 'Electronics') and (`Total Amount` > 100)"

df_clean.query(condition)
```
