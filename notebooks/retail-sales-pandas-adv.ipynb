## Data Transformation

### Create New Columns

New columns were created from existing variables in order to generate more useful metrics and improve the analytical capabilities of the dataset.

These transformations help simplify comparisons between customers, spending behavior, and product pricing levels.

---

#### Normalize Sales

The Total Amount column was normalized using Min-Max normalization to scale values between 0 and 1.

$x' = \frac{x − x_{min}}{x_{max} - x_{min}} $

```python
df_clean['Total Amount Normalized'] = (
    (df_clean['Total Amount'] - df_clean['Total Amount'].min()) /
    (df_clean['Total Amount'].max() - df_clean['Total Amount'].min())
)
```

```python
df_clean[[
    'Transaction ID',
    'Total Amount',
    'Total Amount Normalized'
]].head()
```

**Insight**

Normalization makes values easier to compare by placing all transaction amounts within the same scale. This technique is especially useful for future machine learning and predictive modeling tasks.

#### Age Group

Customers were divided into age ranges to simplify demographic analysis.

```python
df_clean['Age Group'] = pd.cut(
    df_clean['Age'],
    bins = [0, 25, 35, 50, 100],
    labels = ['Young adult', 'Adult', 'Middle age', 'Senior']
)
```

```python
df_clean[[
    'Age',
    'Age Group',
    'Product Category',
    'Total Amount'
]].head()
```

**Insight**

Grouping customers by age makes it easier to identify spending patterns and purchasing behavior across different demographic groups.

#### Spending Category

Transactions were classified into spending levels based on the `Total Amount` column.

```python
df_clean['Spending Category'] = pd.cut(
    df_clean['Total Amount'],
    bins = [0, 100, 500, df_clean['Total Amount'].max()],
    labels = ['Low', 'Medium', 'High'],
    include_lowest = True
)
```

```python
df_clean[[
    'Total Amount',
    'Spending Category'
]].head()
```

**Insight**

This categorization helps distinguish low, medium, and high spending transactions, allowing easier comparison of customer purchasing behavior.

#### Price Tier

Products were grouped into pricing levels according to their unit price.

```python
df_clean['Price Tier'] = pd.cut(
    df_clean['Price per Unit'],
    bins = [0, 50, 200, df_clean['Price per Unit'].max()],
    labels = ['Budget', 'Standard', 'Premium'],
    include_lowest = True
)
```

```python
df_clean[[
    'Price per Unit',
    'Price Tier'
]].head(10)
```

**Insight**

Price segmentation helps identify the distribution of low-cost, mid-range, and premium products within the dataset.

It also supports future comparisons between pricing levels and customer spending behavior.

### Grouping and Aggregation

#### Sales By Month (`sum`)

Monthly sales totals were calculated to identify sales trends throughout the year.

```python
df_clean['Month Number'] = df_clean['Date'].dt.month
df_clean['Month'] = df_clean['Date'].dt.month_name()
```

```python
months_order = [
    'January', 'February', 'March', 'April',
    'May', 'June', 'July', 'August',
    'September', 'October', 'November', 'December'
]

df_clean['Month'] = pd.Categorical(
    df_clean['Month'],
    categories = months_order,
    ordered = True
)
```

```python
df_clean[[
    'Transaction ID',
    'Product Category',
    'Total Amount',
    'Month',
    'Month Number'
]].head()
```

```python
monthly_sales = df_clean.groupby('Month')['Total Amount'].sum()
```

```python
monthly_sales
```

**Observation**

May presented the highest total sales amount, while September recorded the lowest sales performance.

**Insight**

This analysis helps identify the months with the highest and lowest sales performance. Monthly aggregation is commonly used in retail analysis to detect seasonal trends and variations in customer purchasing behavior.

#### Average Spending By Age Group (`mean`)

The average transaction amount was calculated for each age group to compare customer spending behavior across different demographic segments.

```python
df_clean.groupby('Age Group')['Total Amount'].mean().round(2)
```

**Observation**

Young adults presented the highest average spending, while seniors showed the lowest average transaction values.

**Insight**

The results show how spending behavior changes across age groups. Some groups may present higher average purchases, indicating stronger purchasing power or different buying preferences.

#### Transactions By Month (`count`)

The total number of transactions per month was calculated to analyze transaction frequency over time.

```python
df_clean.groupby('Month')['Transaction ID'].count()
```

**Observation**

May recorded the highest number of transactions, while September presented the lowest transaction activity.

**Insight**

This analysis helps identify periods with higher customer activity and allows comparison between transaction volume and total sales performance.

#### Spending Variability by Category (`std`)

The standard deviation of transaction amounts was calculated for each product category.

```python
df_clean.groupby('Product Category')['Total Amount'].std().round(2)
```

**Observation**

All product categories presented similar levels of spending variability, indicating relatively consistent dispersion in transaction amounts across categories.

**Insight**

Standard deviation measures how dispersed transaction values are within each category.

$\sigma = \sqrt{ \frac{1}{N} \sum_{i=1}^{N} (x_i - \mu)^2 }$


Higher values indicate greater variability in customer spending, while lower values suggest more consistent purchasing patterns.

#### Average Quantity by Product Category (`mean`)

The average quantity purchased was calculated for each product category.

```python
df_clean.groupby('Product Category')['Quantity'].mean().round(2)
```

**Observation**

The average quantity purchased was very similar across all product categories, with Clothing showing a slightly higher average quantity.

**Insight**

This analysis helps identify which categories are usually purchased in larger quantities and provides additional insight into customer buying behavior.

### Custom Analysis with `apply()`

#### Deviation from Group Mean

The deviation of each transaction from the average value of its product category was calculated.

```python
df_clean['Deviation from Group Mean'] = (
    df_clean.groupby('Product Category')['Total Amount']
    .apply(lambda x: x - x.mean())
    .reset_index(level = 0, drop = True) # put the indices back in alignment with the dataframe
)
```

```python
df_clean[[
    'Customer ID',
    'Product Category',
    'Total Amount',
    'Deviation from Group Mean'
]].head()
```

**Observation**

Positive deviation values represent transactions above the category average, while negative values indicate transactions below the average.

The calculation follows the formula:

$x_i = \bar{x}_{group}$


**Insight**

Positive values indicate transactions above the category average, while negative values represent transactions below the average.

This type of analysis helps identify unusually high or low purchases within each product category.

#### Detect High-Value Transactions

Transactions were classified according to whether their value was above or below the overall average sales amount.

```python
mean_sales = df_clean['Total Amount'].mean()
```

```python
df_clean['High Value Transaction'] = df_clean['Total Amount'].apply(
    lambda x: 'Yes' if x > mean_sales else 'No'
)
```

```python
df_clean[[
    'Customer ID',
    'Product Category',
    'Total Amount',
    'High Value Transaction'
]].head()
```

**Observation**

Transactions with amounts above the overall average sales value were classified as high-value transactions.

**Insight**

This classification introduces a simple benchmarking approach that helps distinguish higher-value purchases from regular transactions.

It can also support future analysis related to customer segmentation and spending behavior.
